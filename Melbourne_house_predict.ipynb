```python
from sklearn import preprocessing
#import lable encoder as lab_enc
lab_enc=preprocessing.LabelEncoder()
#SUburb have 314 value we can encode it by frequency encoder
freq_sub = round(100*mbl.Suburb.value_counts()/len(mbl.index), 2)
mbl['Suburb'] = mbl['Suburb'].apply(lambda x: freq_sub[x])
```

```python
# room are not needed to be preprocessed
#type h,u,t are converted in lable by lable encoder
mbl['Type']=pd.Series(lab_enc.fit_transform(mbl['Type']))
#methord
mbl['Method']=pd.Series(lab_enc.fit_transform(mbl['Method']))
#SelleG have a total of 263 values
freq_sellerg = round(100*mbl.SellerG.value_counts()/len(mbl.index), 2)
mbl['SellerG'] = mbl['SellerG'].apply(lambda x: freq_sellerg[x])
```

```python
#Distance
streets = []
for addr in mbl["Address"]:
    address = addr.split(' ')
    # street_general_area = address[1:]
    street_general_area = address[-1]
    streets.append(street_general_area)

mbl["Address"] = streets
mbl["Address"] = pd.Series(lab_enc.fit_transform(mbl["Address"]))
```

```python
#Postcode
freq_postcode = round(100*mbl.Postcode.value_counts()/len(mbl.index), 2)
mbl['YearBuilt']=pd.Series(lab_enc.fit_transform(mbl['YearBuilt']))
#councilArea
freq_council = round(100*mbl.CouncilArea.value_counts()/len(mbl.index), 2)
mbl['CouncilArea'] = mbl['CouncilArea'].apply(lambda x: freq_council[x])
#latitude, longitude
#region name
mbl['Regionname']=pd.Series(lab_enc.fit_transform(mbl['Regionname']))
```

```python
X = mbl.drop('Price',axis=1)
y = mbl['Price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=1)
```

```python
sc =StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
reg = LinearRegression()
reg.fit(X_train, y_train)
print(reg.coef_)
print(reg.intercept_)
```

```python
coef = pd.DataFrame(data =reg.coef_, index= X.columns, columns=['Coefficients'] )
coef
```

```python
pred = reg.predict(X_test)
```

```python
plt.scatter(y_test,pred)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test,pred))
print('MSE:', metrics.mean_squared_error(y_test,pred))
print('RMsE:', np.sqrt(metrics.mean_squared_error(y_test,pred)))
```

```python
print('R-squared score (training): {:.3f}'
     .format(reg.score(X_train, y_train)))
print('R-squared score (test): {:.3f}'
     .format(reg.score(X_test, y_test)))
```

```python
from sklearn.linear_model import Ridge

linridge = Ridge(alpha=20.0).fit(X_train, y_train)


print('ridge regression linear model intercept: {}'
     .format(linridge.intercept_))
print('ridge regression linear model coeff:\n{}'
     .format(linridge.coef_))
print('R-squared score (training): {:.3f}'
     .format(linridge.score(X_train, y_train)))
print('R-squared score (test): {:.3f}'
     .format(linridge.score(X_test, y_test)))
print('Number of non-zero features: {}'
     .format(np.sum(linridge.coef_ != 0)))
```

```python
from sklearn.linear_model import Lasso

linlasso = Lasso(alpha=2.0, max_iter = 10000).fit(X_train, y_train)


print('lasso regression linear model intercept: {}'
     .format(linlasso.intercept_))
print('lasso regression linear model coeff:\n{}'
     .format(linlasso.coef_))
print('Non-zero features: {}'
     .format(np.sum(linlasso.coef_ != 0)))
print('R-squared score (training): {:.3f}'
     .format(linlasso.score(X_train, y_train)))
print('R-squared score (test): {:.3f}\n'
     .format(linlasso.score(X_test, y_test)))
print('Features with non-zero weight (sorted by absolute magnitude):')
```

```python
from sklearn.neighbors import KNeighborsRegressor


knnreg = KNeighborsRegressor(n_neighbors = 5).fit(X_train, y_train)

print(knnreg.predict(X_test))
print('R-squared test score: {:.3f}'
     .format(knnreg.score(X_test, y_test)))
```
